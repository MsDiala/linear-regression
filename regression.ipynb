```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df_cars = pd.read_csv('car data.csv')
#Show the info of the data set.
df_cars.info()
```

```python
#Show the first five rows of the data set.
df_cars.head(5)
```

```python
#Show the description 

df_cars.describe()
```

```python

x = df_cars["Present_Price"].values
y = df_cars["Selling_Price"].values
```

```python
X=x.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=100)
```

```python
%matplotlib inline
plt.scatter(x_train,y_train,color='purple')
plt.xlabel('New Car Price')
plt.ylabel('Used Car Price')
plt.title('Sale Data')
plt.show()
```

```python
x_train
```

```python
value = LinearRegression()
value.fit(x_train,y_train)
```

```python
y_predict =value.predict(x_test)
```

```python
y_predict
```

```python
y_test
```

```python
print(f"Train accuracy {round(value.score(x_train,y_train)*100,10)} %")
print(f"Test accuracy {round(value.score(x_test,y_test)*100,10)} %")
```

```python
plt.scatter(x_train,y_train,color='purple')
plt.plot(x_test,y_predict)
plt.xlabel("New Car Price")
plt.ylabel("Used Car Price")
plt.title("Sale Data")
plt.plot
```
